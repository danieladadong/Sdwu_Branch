```python
import jieba
words = jieba.lcut(str(test.get('content')))
```

```python
words
```

```python
words = list(set(words))
```

```python
count = {}
for word in words:            #  使用 for 循环遍历每个词语并统计个数
    if len(word) < 2:          # 排除单个字的干扰，使得输出结果为词语
        continue
    else:
        count[word] = count.get(word, 0) + 1
exclude = ["可以", "一起", "这样",'...']
for key in list(count.keys()):     # 遍历字典的所有键，即所有word
    if key in exclude:
        del count[key]                  #  删除字典中键为无关词语的键值对
lista = list(count.items()) 
```

```python
lista
```

```python
lista.sort(key=lambda x: x[1], reverse=True)
```

```python
la = []
for i in range(5):  #   此处统计排名前五的单词，所以range(5)
    word, number = lista[i]
    la.append(lista[i][0])
    dict1 = {'1,Dec':la}
    print("关键字：{:-<10}频次：{:+>8}".format(word, number))
```
